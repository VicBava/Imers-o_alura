```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df_principal = pd.read_excel("/content/Cópia de Imersão Python - Tabela de ações_Aula 2.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Cópia de Imersão Python - Tabela de ações_Aula 2.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Cópia de Imersão Python - Tabela de ações_Aula 2.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Cópia de Imersão Python - Tabela de ações_Aula 2.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (fundação ou início de operações)':'Idade'}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (anos)':'Ano_de_fundacao'}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Ano_de_fundacao'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
data = {'Segmento': ['Alimentos', 'Aviação', 'Banco', 'Construção', 'Educação', 'Energia', 'Holding', 'Imobiliário', 'Logística', 'Meios de Pagamento', 'Mineração', 'Moda', 'Papel e Celulose', 'Petróleo', 'Química', 'Saúde', 'Shopping Centers', 'Siderurgia', 'Tecnologia', 'Telecomunicações', 'Varejo'],
        'Variacao_rs': [407833683.09, 65452205.55, 4636679965.25, 37525872.38, 72295838.99, 1209821623.57, 416092244.42, 18068446.61, 233902674.79, 43657683.38, 4940442965.55, 41021792.09, 722946282.71, 11336484517.00, 69054317.64, 453917907.01, 117732680.08, 489935930.93, 15598886.65, 292938114.42, 237187009.24]}

# Parâmetros
limite_superior = 700000000  # Definindo um novo limite superior para agrupar mais segmentos
Segmentos_agrupados = []
Variacao_rs_agrupadas = []
total_outros = 0

# Agrupamento de segmentos
for segmento, valor in zip(data['Segmento'], data['Variacao_rs']):
    if valor >= limite_superior:
        Segmentos_agrupados.append(segmento)
        Variacao_rs_agrupadas.append(valor)
    else:
        total_outros += valor

# Verificação e inclusão da categoria 'Outros'
if total_outros > 0:
    Segmentos_agrupados.append('Outros')
    Variacao_rs_agrupadas.append(total_outros)

# DataFrame
df = pd.DataFrame({'Segmento': Segmentos_agrupados, 'Variacao_rs': Variacao_rs_agrupadas})

# Gráfico de pizza
plt.figure(figsize=(10, 6))
plt.pie(df['Variacao_rs'], labels=df['Segmento'], autopct=lambda p: f'R$ {int(p * sum(Variacao_rs_agrupadas) / 100):,.0f}', startangle=140)
plt.title('Variação Reais por Segmento')
plt.axis('equal')
plt.show()
```
